3. Given a recurrence relation, solve it using the recursive tree approach:
a. T(n) = 2T(n-1) +1
b. T(n) = 2T(n/2) + n

a. T(n) = 2T(n-1) +1

T(n)
|
2T(n-1) + 1
|         \
T(n-2)     T(n-2) + 1
|          |        \
T(n-3)    T(n-3)    T(n-3) + 1
...

Each level of the tree represents a different value of n, and the number of branches at each level is determined by the coefficient in front of T(n−1) in the recurrence relation.

The recurrence relation for each level of the tree:

Level 0 (root): T(n)
Level 1 (children of root): 2T(n−1)+1
Level 2 (children of level 1): 2 ** 2 T(n−2)+2+1
Level 3 (children of level 2): 2 ** 3T(n−3)+2**2+2+1
...

Now, let's sum up these expressions over all levels up to level n:

T(n)= 2 ** k T(n-k)+2**k-1...+2+1

At each level, there are 2** k nodes (subproblems) where k is the level number. The work done at each level is 
n times the number of nodes at that level. So, the total work done at a level is n×2** k. 

Now, let's sum up the work done at all levels:

T(n)=n⋅2** 0+n⋅2** 1+n⋅2** 2+…+n⋅2** logn base 2

This is a geometric series, and its sum is given by the formula for the sum of a geometric series:

S=a⋅ 1−r** n/ 1−r 

where a is the first term, r is the common ratio, and n is the number of terms.

In our case, a=n, r=2, and n is the number of levels,which is 
logn base 2.

T(n)=n . 1-2 ** log n+1 base 2 / 1-2

T(n)=n . 1-2.2 ** log n base 2 / -1

T(n)=n. (2 ** logn base 2 -1)

Since, 2 ** logn base 2 = n, we get :

T(n)=n ** 2-n
